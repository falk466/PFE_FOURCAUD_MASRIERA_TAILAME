```python
def getTextPDF(PDFpath):
    pdfFileObj = open(PDFpath, 'rb') 
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj) 

    pageObj = pdfReader.getPage(0)
    pdf = pageObj.extractText()

    for i in range(1, 3):
        pages = pdfReader.getPage(i)
        pdf += pages.extractText()
    pdfFileObj.close()
    return (pdf)
```

```python
def cleanKeywords(keyword):
    keyword = "".join(keyword)
    
    keyword = re.sub('Keywords', '', keyword)
    keyword = re.sub('Introduction', '', keyword)
    keyword = re.sub('abstract', '', keyword)
    keyword = re.sub('Background', '', keyword)
    keyword = re.sub('Key Words', '', keyword)
    keyword = re.sub('KEY WORDS', '', keyword)
    
    regex = re.compile(r'[0-9a-z_-]+[A-Z]+[0-9a-z_-]+')
    if (regex.findall(keyword)):
        temp = keyword.split(" ")
        for i in range (len(temp)):
            temp[i] = re.sub(r'([A-Z]+[0-9a-z_]+)', r',\1', temp[i])
        keyword = " ".join(temp)
        
    keyword = re.sub(';', ',', keyword)
    keyword = re.sub('[\s,]{2,20}', ',', keyword)
    keyword = keyword.split(',')
    keyword.pop(0)
    return keyword
```

```python
from collections import defaultdict

def create_bag_of_words(docs):
    # Initialiser un dictionnaire pour stocker le nombre d'occurrences de chaque mot
    word_counts = defaultdict(int)
    
    # Boucle à travers chaque document
    for doc in docs:
        # Boucle à travers chaque mot dans le document
        for word in doc:
            # Incrémenter le compteur du mot
            word_counts[word] += 1
    
    # Initialiser une liste pour stocker les vecteurs de mots
    bag_of_words = []
    
    # Boucle à travers chaque document
    for doc in docs:
        # Initialiser un vecteur pour stocker les occurrences de chaque mot dans le document
        doc_vector = [0] * len(word_counts)
        
        # Boucle à travers chaque mot dans le dictionnaire
        for i, (word, count) in enumerate(word_counts.items()):
            # Si le mot est présent dans le document, enregistrer son nombre d'occurrences
            if word in doc:
                doc_vector[i] = doc.count(word)
        
        # Ajouter le vecteur du document à la liste des vecteurs de mots
        bag_of_words.append(doc_vector)
    
    return bag_of_words, word_counts.keys()

# Exemple d'utilisation
docs = [["pain", "beurre", "sucre"], ["pain", "beurre", "lait"], ["pain", "oeufs", "lait"]]
bow, words = create_bag_of_words(docs)
print(bow)
print(words)
```

```python
def preprocess_text(text):
    # Lowercase the text
    text = text.lower()
    # Replace "key word" and "keywords" with "keywords"
    text = re.sub(r"key word(s?)", "keywords", text)
    # Replace "key" and "word" with "keywords"
    text = re.sub(r"key\W+word(s?)", "keywords", text)
    # Separate the keyword from its list of words
    text = re.sub(r"keywords\s*:", "keywords:", text)
    return text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

def cluster_texts(keywords_str, num_clusters=3):
    vectorizer = TfidfVectorizer(stop_words='english')
    X = vectorizer.fit_transform(keywords_str)
    
    scores = []
    for ngram_range in [(1,1), (1,2), (1,3)]:
        for max_df in [0.5, 0.75, 1.0]:
            for min_df in [1, 2, 3]:
                for use_idf in [True, False]:
                    vectorizer = TfidfVectorizer(stop_words='english', ngram_range=ngram_range, max_df=max_df, 
                                                 min_df=min_df, use_idf=use_idf)
                    X = vectorizer.fit_transform(keywords_str)
                    km = KMeans(n_clusters=num_clusters)
                    km.fit(X)
                    labels = km.labels_
                    print(labels)
                    score = silhouette_score(X, km.labels_)
                    scores.append((score, ngram_range, max_df, min_df, use_idf))
                
    best_score, best_ngram_range, best_max_df, best_min_df, best_use_idf = max(scores, key=lambda x: x[0])
    print("Best parameters: ngram_range = {}, max_df = {}, min_df = {}, use_idf = {},  with silhouette score = {}".format(
                            best_ngram_range, best_max_df, best_min_df, best_use_idf, best_score))
    
    return

keywords_str = T
cluster_texts(T, num_clusters=3)
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score, adjusted_rand_score
from sklearn.feature_extraction.text import TfidfVectorizer

df = pd.read_csv(r'Keywordsepare.csv') 

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['Keywords'])

agg = AgglomerativeClustering(n_clusters=2, linkage='complete')
agg.fit(X.toarray())

print("Les 10 meilleurs mots par cluster:")
order_centroids = agg.children_
terms = vectorizer.get_feature_names()
true_k=2
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print

print("\n")
print("Prediction")

pred = agg.labels_
print(pred)

silhouette = silhouette_score(X, pred, metric='euclidean')
print(silhouette)

# Calcul des autres métriques
calinski = calinski_harabasz_score(X.toarray(), pred)
davies_bouldin = davies_bouldin_score(X.toarray(), pred)


# Affichage des résultats
print("Calinski-Harabasz score: {:.3f}".format(calinski))
print("Davies-Bouldin score: {:.3f}".format(davies_bouldin))
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import calinski_harabasz_score

# Générer des données aléatoires
X, y = make_blobs(n_samples=1000, centers=4, n_features=2, random_state=42)

# Initialiser les valeurs de k pour le nombre de clusters
k_values = range(2, 10)

# Calculer le score Calinski-Harabasz pour chaque valeur de k
ch_scores = []
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    ch_score = calinski_harabasz_score(X, kmeans.labels_)
    print(ch_score)
    ch_scores.append(ch_score)

# Afficher le graphique des scores Calinski-Harabasz
plt.plot(k_values, ch_scores, 'bx-')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Score Calinski-Harabasz')
plt.title('Optimisation du nombre de clusters')
plt.show()
```
